```python
def total_word_count(text):
    if text:
        return len(text.split())
    else:
        return 0

def percentage_complex_word(text):
    if text:
        word_count = len(text.split())
        complex_word_count = complex_word_count(text)
        return complex_word_count/word_count
    else:
        return 0

def complex_word_count(text):
    if text:
        words = text.split()
        count = 0
        for word in words:
            if len(word) >= 6 and not word.lower() in stopwords.words("english"):
                count += 1
        return count
    else:
        return 0

def AverageSentenceLenght(text):
    if text:
        sentence_list = nltk.sent_tokenize(text)
        return sum(len(sentence.split()) for sentence in sentence_list) / len(sentence_list)
    else:
        return 0

def positive_score(text):
    if text:
        return TextBlob(text).sentiment.polarity
    else:
        return 0

def negative_score(text):
    if text:
        return TextBlob(text).sentiment.polarity
    else:
        return 0

def polarity_score(positive_score, negative_score):
    if positive_score is not None and negative_score is not None:
        return positive_score - negative_score
    else:
        return None
```

```python
corps = []
for url in URLS:
  page = requests.get(url, headers={"User-Agent": "XY"})  
  soup = BeautifulSoup(page.text, 'html.parser')
  # get title
  title_tag = soup.find("h1", attrs={'class': 'entry-title'})
  if title_tag:
      title = title_tag.get_text()
  else:
      title = None

  # get article text
  text_tag = soup.find(attrs={'class': 'td-post-content'})
  if text_tag:
      text = text_tag.get_text()
      # break into lines and remove leading and trailing space on each
      lines = (line.strip() for line in text.splitlines())
      # break multi-headlines into a line each
      chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
      # drop blank lines
      text = '\n'.join(chunk for chunk in chunks if chunk)
  else:
      text = None

  corps.append(text)
```
